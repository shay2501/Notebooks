```python
#import necessary dependencies
import numpy as np
import pandas as pd
```

```python
#load and explore our dataset
df = pd.read_csv('After2000USTempByCity.csv')
df.head()
```

```python
#examine dataset columns
df.columns
```

```python
#rename the unnamed column
df = df.rename(columns={'Unnamed: 0':'recordid'})
df
```

```python
#create a second dataset with just the records from Chicago
df2 = df.loc[df['City'] == 'Chicago']
#take the second dataset and set the city to Rockford 
#we are creating a secondary dataset for example purposes only
df2.loc[:,'City'] = 'Rockford'
df2
```

```python
#adjust Rockford temp to be 2 deg higher than Chicago
df2.loc[:,'AverageTemperature'] += 2
df2
```

```python
#append the Rockford data to the rest of the city data
df = df.append(df2)
```

```python
#see the mean temp for each city, notice Rockford is present
#notice Rockford AverageTemperature is 2 degrees higher than Chicago
df.groupby('City').mean()
```
