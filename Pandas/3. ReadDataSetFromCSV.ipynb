```python
#import necessary dependencies
import numpy as np
import pandas as pd
```

```python
#load and explore our dataset
df = pd.read_csv('After2000USTempByCity.csv')
df.head(10)
```

```python
#set our columns and look at the last 5 records in our dataframe
df.columns = ['RecordId', 'Date', 'City', 'Country', 'AverageTemperature']
df.tail()
```

```python
#find some basic info about our dataframe
df.info()
```

```python
#look at the datatypes for each column
df.dtypes
```

```python
#see the shape, (# of rows, # of columns) in the dataset
df.shape
```

```python
#explore what the city data looks like
df['City']
```

```python
#get some basic statistics about the dataset
df.describe()
```

```python
#sort the dataset columns
df= df.sort_index(axis=1)
```

```python
#slice the dataset getting the first three rows
df[0:3]
```

```python
#set the dataframe index to the city column
df.index = df['City']
df
```

```python
#slice the dataset to get the city of the first two rows
df.iloc[0:2, 1]
```

```python
#slice the dataset getting all rows and just the City and Temp columns
df.loc[:, ['City', 'AverageTemperature']]
```

```python
#slice the dataset getting some rows and columns from the middle
df.iloc[400:410,1:3]
```

```python
#create a new column, 'X' and assign it the value of 2 for all rows
df.loc[:, 'X'] = 2
df
```

```python
#get all unique values for City in the dataset
df['City'].unique()
```

```python
#create a dictionary of city-state
states = {'Chicago': 'Illinois', 'New York': 'New York', 'Los Angeles': 'California'}
#Create the state column and map values from the dictionary
df['State'] = df['City'].map(states)
df.loc[df['City'] == 'Los Angeles']
```

```python
#create a column 'Y' and set every 3rd record the value of True
df.loc[df['RecordId']%3 == 0, 'Y'] = True
df
```

```python
#explore the dataset looking for null and NaN values
pd.isna(df)
```

```python
#reset the index back to the original integer index
#drops the index instead of inserting it to the dataframe 
df.reset_index(drop=True)
```

```python
#create a multi-column index modifiying the dataframe in place
df.set_index(['State', 'City'], inplace=True)
```

```python
#slice the dataframe for Los Angeles, CA
df.loc[['California', 'Los Angeles']]
```

```python
#save the dataframe to a csv file
df.to_csv('heirarchy_citystate.csv')
```

```python
#manipulate the Country column to be all upper case
df.loc[:,'Country'].str.upper()
```

```python
#display the mean for every column
df.mean()
```
